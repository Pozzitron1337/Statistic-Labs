Перевірку статистичних гіпотез вести при рівні значимості gamma=0.05 . Кожне з наступних чотирьох завдань виконувати для n=1000,n=10000, n=100000,  та . Користуючись перетворенням Y_i=F(X_i;lambda),i=1,...,n перевіряти на рівномірність випадкові величини {Y_i} (лише перші три завдання).

<b>Завдання 1:</b> 
за допомогою критерія Колмогорова перевірити гіпотези:
<p>a) H_0: X_i ~ F(u,1), коли насправді X_i~F(u.1) ;</p>
<p>b) H_0: X_i ~ F(u,1) , коли насправді X_i~F(u,1.3).</p>

```python
def F(u,l):
    return 1-np.exp(np.negative(np.multiply(u,l)))
    
def D(x,l,n):
    x_sorted=np.sort(x)
    D_n=0
    D_k=lambda y: np.maximum(y-(k-1)/n,k/n-y)
    for k in range(n):
        new_D_k=D_k(F(x_sorted[k],l))
        if D_n<new_D_k:
            D_n=new_D_k
    return D_n

def kolmogorov(x,z,l):
    n=len(x)
    D_n=D(x,l,n)
    criteria=np.sqrt(n)*D_n
    print(f'l = {l}')
    print(f'n = {n}')
    print(f'Приблизні значення вибірки x= {x[0:5]}')
    print(f'D_n(x) = {D_n}')
    print(f'sqrt(n)*D_n(x) = {criteria}')
    print(f'z= {z}')
    if criteria<z:
        print(f'Висновок:статистичні дані не суперечать гіпотезі H_0\n')
    else:
        print(f'Висновок:слід прийняти альтернативну гіпотезу H_1\n')

def taskA():
    l=1
    z=1.36 #z=K^{-1}(1-0.05)
    x1=np.random.exponential(scale=1,size=1000)
    kolmogorov(x1,z,l)
    x2=np.random.exponential(scale=1,size=10000)
    kolmogorov(x2,z,l)
    x3=np.random.exponential(scale=1,size=100000)
    kolmogorov(x3,z,l)

def taskB():
    l=1.3
    z=1.36 #z=K^{-1}(1-0.05)
    x1=np.random.exponential(scale=1,size=1000)
    kolmogorov(x1,z,l)
    x2=np.random.exponential(scale=1,size=10000)
    kolmogorov(x2,z,l)
    x3=np.random.exponential(scale=1,size=100000)
    kolmogorov(x3,z,l)
    
taskA()
taskB()
```

<b>Завдання 2:</b> за допомогою критерія X^2 перевірити гіпотези:
<p>a) H_0: X_i ~ F(u,1), коли насправді X_i~F(u.1) ;</p>
<p>b) H_0: X_i ~ F(u,1) , коли насправді X_i~F(u,1.3).</p>
<b>Зауваження.</b> Кількість проміжків r обирати з умови r=20*n/1000: 

```python
def F(u,l):
    return 1-np.exp(np.negative(np.multiply(u,l)))

def form_v(x,l,r):
    interval=0
    x_norm=F(x,l)
    left=0
    right=1.0/r
    v=[0]*r
    for i in range(r):
        U_i=x_norm[(left<=x_norm)&(x_norm<right)]
        v[i]=len(U_i)
        left+=1/r
        right+=1/r
    return v

def delta(v,p,n,r):
    d=0.0
    for i in range(r):
        d+=(v[i]-n*p)**2/(n*p)
    return d
        
def chi2(x,l,g):
    r=int(20*len(x)/1000)
    z=st.chi2.ppf(1-g,r-1)
    n=len(x)
    v=form_v(x,l,r)
    p=1/r
    d=delta(v,p,n,r)
    print(f'l = {l}')
    print(f'n = {n}')
    print(f'r = {r}')
    print(f'Приблизні значення вибірки x: {x[0:4]}')
    print(f'Приблизні значення v: {v[0:4]}')
    print(f'Довжина інтервалу p = {p}')
    print(f'delta = {d}')
    print(f'z = {z}')
    if d<z:
        print(f'Статистичні дані не суперечать гіпотезі H_0\n')
    else:
        print(f'Слід прийняти альтернативну гіпотезу H_1\n')
    
def taskA():
    l=1
    g=0.05
    x1=np.random.exponential(scale=l,size=1000)
    chi2(x1,l,g)
    x2=np.random.exponential(scale=l,size=10000)
    chi2(x2,l,g)
    x3=np.random.exponential(scale=l,size=100000)
    chi2(x3,l,g)
    
def taskB():
    l=1.3
    g=0.05
    x1=np.random.exponential(scale=l,size=1000)
    chi2(x1,l,g)
    x2=np.random.exponential(scale=l,size=10000)
    chi2(x2,l,g)
    x3=np.random.exponential(scale=l,size=100000)
    chi2(x3,l,g)

taskA()
taskB()
```

<b>Завдання 3:</b> за допомогою критерія пустих ящиків перевірити гіпотези:
<p>a) H_0: X_i ~ F(u,1), коли насправді X_i~F(u.1) ;</p>
<p>b) H_0: X_i ~ F(u,1) , коли насправді X_i~F(u,1.3).</p>
<b>Зауваження.</b> Кількість проміжків  обирати з умови: p=2, тобто із співвідношення n/r=p випливає, що r=n/2.

<img src="./laplace.jpg">

```python
def F(u,l):
    return 1-np.exp(np.negative(np.multiply(u,l)))

def form_v(x,l,r):
    x_norm=F(x,l)
    left=0
    right=1.0/r
    v=np.array([0]*r)
    for i in range(r):
        U_i=x_norm[(left<=x_norm)&(x_norm<right)]
        v[i]=len(U_i)
        left+=1/r
        right+=1/r
    return v
def form_nu_0(v):
    nu_0=len(v[v==0])
    return nu_0
    
def blank_boxes(x,l):
    n=len(x)
    p=2
    r=int(n/p)
    z=st.norm.ppf(1-0.05)#z=1.6448536269514722 look at the picture
    criteria=r*np.exp(-p)+z*np.sqrt(r*np.exp(-p)*(1-(1+p)*np.exp(-p)))
    v=form_v(x,l,r)
    nu_0=form_nu_0(v)
    print(f'l = {l}')
    print(f'n = {n}')
    print(f'r = {r}')
    print(f'Приблизні значення вибірки x: {x[0:4]}')
    print(f'Приблизні значення v: {v[0:4]}')
    print(f'Довжина інтревалу p = {1/r}')
    print(f'Кількість пустих ящиків nu_0 = {nu_0}')
    print(f'criteria = {criteria}')
    if nu_0<criteria:
        print(f'Статистичні дані не суперечать гіпотезі H_0\n')
    else:
        print(f'Слід прийняти альтернативну гіпотезу H_1\n')
    
def taskA():
    l=1
    x1=np.random.exponential(scale=l,size=1000)
    blank_boxes(x1,l)
    x2=np.random.exponential(scale=l,size=10000)
    blank_boxes(x2,l)
    x3=np.random.exponential(scale=l,size=100000)
    blank_boxes(x3,l)
    
def taskB():
    l=1.3
    x1=np.random.exponential(scale=l,size=1000)
    blank_boxes(x1,l)
    x2=np.random.exponential(scale=l,size=10000)
    blank_boxes(x2,l)
    x3=np.random.exponential(scale=l,size=100000)
    blank_boxes(x3,l)
    
taskA()
taskB()
```

<b>Завдання 4:</b> за допомогою критерія однорідності Смирнова перевірити гіпотези:
<p> a) H_0: X_1=(x_1,...,x_n) ~ F(u,1), X_2=(x_1,...,x_m) ~ F(u,1)
(саме так ці вибірки і генерувались);</p>
<p>b)  H_0: X_1=(x_1,...,x_n) ~ F(u,1), X_2=(x_1,...,x_m) ~ F(u,1)</p>
<p>(насправді: X_1 ~ F(u,1), X_2 ~ F(u,1.3)).</p>
<b>Зауваження.</b> Обирати m=n/2.

```python
def F(u,l):
    return 1-np.exp(np.negative(np.multiply(u,l)))

def D_plus(y,m,l):
    y_sorted=np.sort(y)
    max=y_sorted[0]
    for k in range(m):#k=0,1,...,m-1
        temp=((k+1)/m)-F(y_sorted[k],l)
        if max<temp:
            max=temp
    return max

def D_minus(y,m,l):
    y_sorted=np.sort(y)
    max=y_sorted[0]
    for k in range(m):#k=0,1,...,m-1
        temp=F(y_sorted[k],l)-(k/m)
        if max<temp:
            max=temp
    return max

def smirnov(x,y,z,l):
    n = len(x)
    m = len(y)
    D_p = D_plus(y,m,l)
    D_m = D_minus(y,m,l)
    D=np.maximum(D_p,D_m)
    criteria=z*np.sqrt((1/n)+(1/m))
    print(f'l = {l}')
    print(f'n = {n}')
    print(f'm = {m}')
    print(f'Приблизні значення вибірки x: {x[0:4]}')
    print(f'Приблизні значення вибірки y: {y[0:4]}')
    print(f'D+ = {D_p}')
    print(f'D- = {D_m}')
    print(f'D = {D}')
    print(f'criteria = {criteria}')
    if D<criteria:
        print(f'Статистичні дані не суперечать гіпотезі H_0\n')
    else:
        print(f'Слід прийняти альтернативну гіпотезу H_1\n')
    

def taskA():
    l=1
    z=1.36 #z=K^{-1}(1-0.05)
    n=1000
    m=n//2
    x1=np.random.exponential(scale=l,size=n)
    y1=np.random.exponential(scale=l,size=m)
    smirnov(x1,y1,z,l)
    
    n=10000
    m=n//2
    x2=np.random.exponential(scale=l,size=n)
    y2=np.random.exponential(scale=l,size=m)
    smirnov(x2,y2,z,l)
    
    n=100000
    m=n//2
    x3=np.random.exponential(scale=l,size=n)
    y3=np.random.exponential(scale=l,size=m)
    smirnov(x3,y3,z,l)
    
def taskB():
    l_1=1
    l_2=1.3
    z=1.36 #z=K^{-1}(1-0.05)
    n=1000
    m=n//2
    x1=np.random.exponential(scale=l_1,size=n)
    y1=np.random.exponential(scale=l_2,size=m)
    smirnov(x1,y1,z,l)
    
    n=10000
    m=n//2
    x2=np.random.exponential(scale=l_1,size=n)
    y2=np.random.exponential(scale=l_2,size=m)
    smirnov(x2,y2,z,l)
    
    n=100000
    m=n//2
    x3=np.random.exponential(scale=l_1,size=n)
    y3=np.random.exponential(scale=l_2,size=m)
    smirnov(x3,y3,z,l)
    
taskA()
taskB()
```
